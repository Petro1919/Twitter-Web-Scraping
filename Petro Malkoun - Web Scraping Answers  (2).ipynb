1- **Twitter EXERCISE:** 

Given the starting user "gatdelapena", retrieve the user name, the date of the accoount creation and the last tweet date and text for each follower. Store them to a database called 'twitter' and into a collection called 'info'. The attributes of the document are: 'screen_name', 'created_at', 'last_tweet_text' and 'last_tweet_created_at'. 


```python
import json
import pymongo
import tweepy

with open('consumer_key.txt', 'r') as f:
    consumer_key =  f.read()
f.closed
with open('consumer_secret.txt', 'r') as f:
    consumer_secret = f.read()
f.closed
with open('access_key.txt', 'r') as f:
    access_key = f.read()
f.closed
with open('access_secret.txt', 'r') as f:
     access_secret = f.read()
f.closed

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)

USER_ID="gatdelapena"
# USER_ID = "777989114964606976"

user = api.get_user(id=USER_ID)
USER_NAME=user.screen_name
print('User ID:', USER_ID)
print('Username:',user.screen_name)
print('User Created at:',user.created_at)
print('Please wait a moment...')
print()

timeline = api.user_timeline(id=USER_NAME,count=1)
twitter= [] 
followers = tweepy.Cursor(api.followers, screen_name=USER_NAME).items()  

try:
    for follower in followers:
        new_tweets = api.user_timeline(screen_name = follower.screen_name,count=1, tweet_mode="extended")
        for tweet in new_tweets:
            tweets = [tweet.created_at,tweet.full_text] 
            date=tweet.created_at          
#         print(tweets)
#         print()
        twitter.append({'screen_name':follower.screen_name,'created_at':follower.created_at,
                        'last_tweet':tweets, 'last_tweet_created_at':date})
        
except tweepy.TweepError:
    print('THANK YOU!')
    print()
    print("Failed to run the command on that user, Skipping...")
print()
print('THE SET OF TWEETS IS:',twitter) 
print()
try:
    with open("credentials.txt", 'r') as f:
        [name,password,url]=f.read().splitlines()
        conn=pymongo.MongoClient("mongodb+srv://{}:{}@{}".format(name,password,url))
    print ("Connected successfully!!!")
except pymongo.errors.ConnectionFailure as e:
        print ("Could not connect to MongoDB: %s" % e) 
        
db = conn['twitter']
collection = db['info']
collection.insert_many(twitter)
print()
print("This Data is successfully stored in Mongodb!!")
```

```python
from selenium import webdriver
import time,os
import pymongo
                                                              #osas b session 5

url = 'https://www.tripadvisor.com/Restaurants' #Poseu el nom de la pàgina web
# path = "/Users/eloi/chromedriver"
path = r"C:\Users\User\Desktop\Term 1 MIBA\Cloud Computing\Session 4\CC_ETL\Web-Scraping-Notebook\chromedriver"
browser = webdriver.Chrome(path)
browser.get(url)
element = browser.find_element_by_class_name("typeahead_input")
element.clear()
time.sleep(1)
s="Sant Cugat"
time.sleep(1)
element.send_keys(s)
element.send_keys(u'\ue007')
time.sleep(3)
names = browser.find_elements_by_xpath("//*[contains(@class,'restaurants-list-ListCell__nameBlock')]")
links = browser.find_elements_by_xpath("//*[contains(@class,'restaurants-list-ListCell__restaurantName')]")
top10 = []  
print("The Top 10 Restaurants in Sant Cugat on TripAdvisor are:")
print()
n=1
while n<11:  
    first = {'name':names[n].text, 'link':links[n].get_attribute('href')}
    top10.append(first)
    print(names[n].text)
    print(links[n].get_attribute('href'))
    print()
    n+=1
   
try:
    with open("credentials.txt", 'r') as f:
        [name,password,url]=f.read().splitlines()
        conn=pymongo.MongoClient("mongodb+srv://{}:{}@{}".format(name,password,url))
    print ("Connected successfully!!!")
except pymongo.errors.ConnectionFailure as e:
        print ("Could not connect to MongoDB: %s" % e) 

db = conn['TripAdvisor']
collection = db['Top 10 Restaurants']
collection.insert_many(top10)
print()
print("This Data is successfully stored in Mongodb!!")
```
